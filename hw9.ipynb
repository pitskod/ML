**Задание 1.** 

Изучите понятие ROC и ROC-AUC. Докажите, что площадь под ROC-кривой численно равна вероятности оценить объект положительного класса выше объекта отрицательного.

**Решение:** 
Возьмем два объекта ($T^i$,$Y^i$) и ($T^j$,$Y^j$) $i \neq j$, где $T^i$,$T^j$ - вероятность отнести объект $i$ и $j$ к классу $1$,
$Y^i$,$Y^j$ - истинные метки класса.

$AUC=\int_{1}^{0}TPR(t)d(FPR(t))$

Выразим TPR и FPR через условные вероятности:
$TPR =  \frac{TP(t)}{TP(t)+FN(t)}=\frac{TP(t)/n}{TP(t)/n+FN(t)/n}=\frac{P(T>t,Y=1)}{P(T>t,Y=1)+P(T \leq t,Y=1)}=\frac{P(T>t,Y=1)}{P(Y=1)}=P(T>t \rvert Y=1)$

$FPR = 1 - \frac{TN(t)}{TN(t)+FP(t)}=\frac{FP(t)}{TN(t)+FP(t)}=P(T>t \rvert Y=0)$ 

$AUC=\int_{1}^{0}TPR(t)d(FPR(t))=\int_{1}^{0}P(T>t \rvert Y=1)d(P(T>t \rvert Y=0))=\int_{1}^{0}P(T>t \rvert Y=1)(-p(t \rvert Y=0))=\int_{0}^{1}P(T>t \rvert Y=1) \ p(t \rvert Y=0)$

Найдем вероятность оценить объект положительного класса выше объекта отрицательного, т.е
оценить вероятность  того, что значение $T^i$ будет больше, чем $T^j$ при условии $Y^i = 1$, $Y^j = 0$.

$P(T^i > T^j \rvert Y^i = 1, Y^j = 0)=$
$\iint\limits_{\{(t^i, t^j):t^i > t^j\}} p(t^i, t^j \rvert Y^i = 1, Y^j = 0) dt^i dt^j=$ [$i$ и $j$ независимы]
$=\iint\limits_{\{(t^i, t^j):t^i > t^j)\}} p(t^i \rvert Y^i = 1 ) \ p(t^j \rvert  Y^j = 0) dt^i dt^j=$
$=\int\limits_{0}^{1} \int\limits_{t^j}^{1} p(t^i \rvert Y^i = 1 ) \ p(t^j \rvert  Y^j = 0) dt^i dt^j=$
$\int_{0}^{1} \left( \int_{t^j}^{1} p(t^i \rvert Y^i = 1 ) dt^i \right) p(t^j \rvert  Y^j = 0)  dt^j=$
$\int_{0}^{1}  P(T^i > t^j \rvert Y^i = 1 ) \ p(t^j \rvert  Y^j = 0)  dt^j=$
$=\int_{0}^{1}  P(T > t \rvert Y = 1 ) \ p(t \rvert  Y = 0)  dt=AUC$

**Задание 2.** 

Приведите пример задачи, когда precision важней recall, когда recall важней precision и когда необходимо находить баланс между этими двумя величинами

**Решение:** 

$Precision =  \frac{TP}{TP+FP}$

$Recall =  \frac{TP}{TP+FN}$

$Precision(точность)$ можно интерпретировать как долю объектов, названных классификатором положительными и при этом действительно являющимися положительными, а $recall(полнота)$ показывает, какую долю объектов положительного класса из всех объектов положительного класса нашел алгоритм.
$Recall$ демонстрирует способность алгоритма обнаруживать данный класс вообще, а $precision$ — способность отличать этот класс от других классов.

Рассмотрим задачу классификации как задачу выявления объектов положительного класса и в качестве положительного класса рассматривать "плохое" событие. 

Оценивать качество модели с помощью точности следует тогда, когда необходимо снизить количество ложных(ложноположительный) срабатываний, т.е. уменьшить ошибку первого рода,снизить количество "ложных тревог"(говорим, что "плохое" событие наступило, когда на самом деле все "хорошо"). 

С помощью точности можно оценивать качество в слудущих задачах:

1) Диагностики болезней. В первую очередь важно выявить всех больных пациентов, при этом, возможно, включив в их число и небольшое количество здоровых пациентов.


Оценивать качество модели с помощью полноты следует тогда, когда необходимо снизить количество ложноотрицательных  срабатываний, т.е. уменьшить ошибку второго рода, снизить количество "пропусков событий"(говорим, что все "хорошо",  когда на самом деле все "плохо").

С помощью полноты можно оценивать качество в слудущих задачах:

1) Идентификация по отпечатку пальца. Важно не допустить, чтобы система пропустила не того человека, полнотой можно незначительно пожертвовать - нет ничего сложного чтобы еще раз поднести палец к датчику.

2) Системы распознавания "свой"/"чужой" самолет. Здесь ни в коем случае нельзя пропустить "чужой" самолет.

3) Выявление мошенничества по банковским картам. В этой задаче полнота важнее точности, т.к. для этой задачи крайне важно не пропустить мошеннические действия(1-2% из всех транзакций) и можно пожертвовать точностью (позже проверяются вручную и ложные срабатывания отсеятся).

Таким образом, полнота важнее в задачах, где стоимость ошибочного срабатывания высока.

Находить баланс между точностью и полнотой разумно в задачах, где цена ложногоотрицательного и ложноположительного срабатывания приблизительно равна.

1) Таргетинговая реклама. Полнота отвечает за то, чтобы показать рекламу как можно большему количеству пользователей, а точность - показать наиболее заинтересованным и которые перейдут по ссылке.

2) Отток клиентов. Важно обзвонить как можно больше клиентов(полнота), чтобы удержать клиента, Однако необходимо точно определить этих клиентов, чтобы сэкономить время операторов(точность).
