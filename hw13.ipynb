**Задание 1.** 

Доказать, что $M_{SOA}(H) = Ldim(H)$

**Решение:** 

Докажем сдедующие утверждения:

1) для SOA выполняется $M_{SOA}(H) \leq Ldim(H)$

2) $M_{A} \geq Ldim(H)$

Из 1)-2) будет следовать, что $M_{SOA}(H) = Ldim(H)$

1) Понятно, что $Ldim(V_{t})$ не может возрастать, т.к. класс гипотез не увеличивается по ходу алгоритма. Всякие раз когда алгоритм SOA делает ошибку $Ldim(V_{t+1}) \leq Ldim(V_{t})-1$. Докажем это от противного, т е предполагая что  $Ldim(V_{t+1})=Ldim(V_{t})$ . Но тогда получается, что $Ldim(V_{t}^{r})=Ldim(V_{t})$ для $r=1$ и $r=0$ .Тогда можно построить раскрашенное дерево глубины  $Ldim(V_{t})+1$. Получили противоречие. Т. е справедливо $Ldim(V_{t+1}) \leq Ldim(V_{t})-1$ что означает следующее: после каждой ошибки $Ldim(V)$ уменьшается хотя бы на единицу, а значит, количество ошибок алгоритма подчиняется неравенству  $M_{SOA}(H) \leq Ldim(H)$. 

2) Пусть $T = Ldim(H)$ и последовательность вершин $(v_{1},v_{2},...,v_{2^{T}-1})$ удовлетворяющих определению $Ldim$ (существует разукрашиваемое дерево высоты $T$). Если природа для всех $t = 1,...,T$ будет указывать для $v_{i_{t}}$ объекта метку, противоположную предсказанию алгоритма, алгоритм ошибется ровно $T$ раз. Т.е. не существует алгоритма у которого $M_{A} < Ldim(H)$.

**Задание 3.** 

Найти класс $H$, что алгоритм $Consistent$ делает на нём $|H| − 1$ ошибку

**Решение:** 

Рассмотрим $n$ объектов $X=\{x_{1},...x_{n}\}$ таких что класс этих обектов для $x_{1},...x_{n-1}$ равен $0$, для $x_{n}$ равен $1$. 

Рассмотрим также множество гипотез $H=\{h_{i}  |   h_{i}(x_{j})=1, если i=j; h_{i}(x_{j}) = 0, если i\neq j  \}$, $|H|=n$. 

Для того чтобы алгоритм совершил $|H| − 1$ ошибку нужно,  чтобы при каждой ошибке алгоритм отбрасывал ровно одну гипотезу из класса.
Пусть мы выбираем гипотезы случайно и эта случайность соответсвует отсортированным по возрастанию индекса гипотезам.


Единственная гипотеза $h_{n}$ дает правильный ответ на всех объектах.

Пусть объекты также случайным  образом приходят в отсортированном по индексу порядку.  На $i$-ой итерации выбирается $i$-ая гипотеза и проверяется на $i$-ом объекте. И эта же $i$-ая гипотеза исключается из множества далее рассматриваемых гипотез. На последней итерации останется единственная гипотеза $h_{n}=h^{*}$
Общее число ошибок при таком псевдо случайном выборе гипотезы будет равно $|H|−1$.

**Задание 4.** 

Найти класс $H$, что алгоритм $Halving$ делает на нём ровно $log_{2}(|H|)$ ошибок

**Решение:** 

Алгоритм совершит ровно $log_{2}(|H|)$ ошибок, если подобрать такую последовательность и такой класс гипотез, что на каждой итерации алгоритм ошибется и после каждой итерации $V_{t}$ будет уменьшаться вдвое за счет того, что ровно половина гипотез проголосует за положительный класс, а вторая за отрицательный.

Когда возникает такая неоднозначная ситуация, будем предсказывать метку 1. 

Т.к. требуется ровно логарифм ошибок, класс гипотез должен быть размера, кратного степени двойки. 

Рассмотрим множество бинарных векторов длины $n$, $n=2^{m}$. 

Множество гипотез  $H=\{h_{i}=x_{i},i=1,...,n\}$ ($i$-ая координата вектора).
Истинная гипотеза $h^{*}=h_{1}=x_{1}$.

Пусть первым "случайно" приходит объект, у которого на $0,...,n/2$  местах стоят $0$, а на остальных местах $1$.
Тогда  $n/2$ гипотезы $h_{n/2+1},...,h_{n}$ ответять $"1"$,   $n/2$ гипотезы $h_{1},...,h_{n/2}$ ответять $"0"$. В такой ситуации мы говорим, что объект принадлежит классу 1, а значит гипотезы    $h_{n/2+1},...,h_{n}$  будут исключены.

На $i$-ом шаге приходит объект, у которого на $0,...,n/(2^{i})$  местах стоят $0$, а на остальных местах $1$.
Тогда  $n/(2^{i})$ гипотезы $h_{n/(2^{i})+1},...,h_{n/(2^{i-1})}$ ответять $"1"$,   $n/(2^{i})$ гипотезы $h_{1},...,h_{n/(2^{i})}$ ответять $"0"$. В такой ситуации мы говорим, что объект принадлежит классу 1, а значит гипотезы    $h_{n/(2^{i})+1},...,h_{n/(2^{i-1})}$ будут исключены.

Алгоритм придет к гипотезе $h^{*}$, сделав ровно $log_{2}(|H|)$ ошибок.
